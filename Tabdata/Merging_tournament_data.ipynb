```python
import pandas as pd
import numpy as np
import os
```

```python
# List only files in the current directory
files = [f for f in os.listdir('.') if os.path.isfile(f)]

for f in files:
    print(f)
    
txt_files = [f for f in files if f.lower().endswith('.txt')]

print(txt_files)
```

```python
def open_excel_comp(txt_file):
    comp_name = txt_file[:-4]
    print(comp_name)
    df = pd.read_excel(comp_name+"_rounds_done.xlsx")
    print(len(df))
    return df, comp_name
    
data = pd.DataFrame()


for i in range(len(txt_files)):
    df, comp_name = open_excel_comp(txt_files[i])
    data = pd.concat([df, data], ignore_index=True)
```

```python
data.to_excel("All_Debates.xlsx", index=False)  # index=False removes the row index
```
