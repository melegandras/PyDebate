```python
import time
import pyperclip
import pandas as pd
import openpyxl
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)  # waits up to 10 seconds for elements
```

```python
# Load workbook and select sheet
wb = openpyxl.load_workbook("Majors_Tournament_Masterlist.xlsx")
ws = wb.active  # or wb['SheetName']

# Extract cell values and hyperlinks
data = []


for row in ws.iter_rows(min_row=2):  # skip header if needed
    row_data = []
    for cell in row:
        if cell.hyperlink:
            row_data.append(cell.hyperlink.target)
        else:
            row_data.append(cell.value)
    data.append(row_data)

# Optional: get headers from first row
headers = [cell.value for cell in ws[1]]

# Create DataFrame
df = pd.DataFrame(data, columns=headers)
df = df.dropna()

### Last bit to solve the crucs

#for i in range(len(df)):
for i in range(100):
    print(i)
    time.sleep(3)
    comp_link = df.iloc[i]["Comp URL"]
    comp_name = df.iloc[i]["Comp longname"]
    print(f"{comp_link} is on {comp_name}")

    driver.get(comp_link)
    time.sleep(2)
    with open(comp_name+".txt", "w", encoding="utf-8") as f:
        pass  # This just creates/truncates the file

    driver.find_element(By.LINK_TEXT, "Team Tab").click()

    copy_button = wait.until(EC.element_to_be_clickable(
        (By.CSS_SELECTOR, "button[title][data-original-title*='CSV format']")  # Adjust if necessary
    ))
    copy_button.click()

    # Wait for clipboard to update
    time.sleep(1)  # Ensure clipboard content is ready

    # Paste from clipboard using pyperclip
    copied_data = pyperclip.paste()

    # Append to txt with newline separator
    with open(comp_name+".txt", "a", encoding="utf-8") as f:
        f.write("\n" + copied_data + "\n")  # Add newline before and after

    print("Copied teamtab content appended to file with separator.")

    driver.find_element(By.LINK_TEXT, "Speaker Tab").click()

    copy_button = wait.until(EC.element_to_be_clickable(
        (By.CSS_SELECTOR, "button[title][data-original-title*='CSV format']")  # Adjust if necessary
    ))
    copy_button.click()

    # Wait for clipboard to update
    time.sleep(1)  # Ensure clipboard content is ready

    # Paste from clipboard using pyperclip
    copied_data = pyperclip.paste()

    # Append to txt with newline separator
    with open(comp_name+".txt", "a", encoding="utf-8") as f:
        f.write("\n" + copied_data + "\n")  # Add newline before and after

    print("Copied speaker tab content appended to file with separator.")

    wait = WebDriverWait(driver, 10)
    dropdown_toggle = wait.until(EC.element_to_be_clickable((By.ID, "roundsDrop")))
    dropdown_toggle.click()

    # Wait for the dropdown items to be visible
    dropdown_items = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".dropdown-menu .dropdown-item")))

    # Extract only the href attributes into a list
    links = [item.get_attribute("href") for item in dropdown_items]

    # Use the list as needed
    #print(links)

    round_links = [link for link in links if "/results/round/" in link]
    print(round_links)

    for i in range(len(round_links)):
        print(i)

        driver.get(round_links[i])
        wait = WebDriverWait(driver, 10)
        driver.find_element(By.LINK_TEXT, "View by Team").click()

        copy_button = wait.until(EC.element_to_be_clickable(
        (By.CSS_SELECTOR, "button[title][data-original-title*='CSV format']")  # Adjust if necessary
        ))
        copy_button.click()

        # Wait for clipboard to update
        time.sleep(1)  # Ensure clipboard content is ready

        # Paste from clipboard using pyperclip
        copied_data = pyperclip.paste()

        # Append to txt with newline separator
        with open(comp_name+".txt", "a", encoding="utf-8") as f:
            f.write("\n" + copied_data + "\n")  # Add newline before and after

        print("Copied rounds content appended to file with separator.")
```
